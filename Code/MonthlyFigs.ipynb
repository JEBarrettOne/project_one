```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from scipy import stats

# Data files
library_covid_data = "../Data/merged_library_covid_outer.csv"

# Read the mouse data and the study results
citywidelib = pd.read_csv(library_covid_data, dtype={'JANUARY': float, 'YTD': float, 
                                                           'YEAR': object, 'ZIP': object})

citywidelib.head()
```

```python
cityinfo2019 = citywidelib.loc[citywidelib['YEAR'] == "2019"]
cityinfo2020 = citywidelib.loc[citywidelib['YEAR'] == "2020"]
cityinfo2021 = citywidelib.loc[citywidelib['YEAR'] == "2021"]
```

```python
grouped19 = cityinfo2019.groupby(['TYPE']).agg({'JANUARY': ['sum', 'mean'], 'FEBRUARY': ['sum', 'mean'], 'MARCH': ['sum', 'mean'], 'APRIL': ['sum', 'mean'], 
                                            'MAY': ['sum', 'mean'], 'JUNE': ['sum', 'mean'], 'JULY': ['sum', 'mean'], 'AUGUST': ['sum', 'mean'], 
                                            'SEPTEMBER': ['sum', 'mean'], 'OCTOBER': ['sum', 'mean'], 'NOVEMBER': ['sum', 'mean'], 
                                             'DECEMBER': ['sum', 'mean']}).round(0)
grouped20 = cityinfo2020.groupby(['TYPE']).agg({'JANUARY': ['sum', 'mean'], 'FEBRUARY': ['sum', 'mean'], 'MARCH': ['sum', 'mean'], 'APRIL': ['sum', 'mean'], 
                                            'MAY': ['sum', 'mean'], 'JUNE': ['sum', 'mean'], 'JULY': ['sum', 'mean'], 'AUGUST': ['sum', 'mean'], 
                                            'SEPTEMBER': ['sum', 'mean'], 'OCTOBER': ['sum', 'mean'], 'NOVEMBER': ['sum', 'mean'], 'DECEMBER': ['sum', 'mean']}).round(0)

grouped21 = cityinfo2021.groupby(['TYPE']).agg({'JANUARY': ['sum', 'mean'], 'FEBRUARY': ['sum', 'mean'], 'MARCH': ['sum', 'mean'], 'APRIL': ['sum', 'mean'], 
                                            'MAY': ['sum', 'mean'], 'JUNE': ['sum', 'mean'], 'JULY': ['sum', 'mean'], 'AUGUST': ['sum', 'mean'], 
                                            'SEPTEMBER': ['sum', 'mean'], 'OCTOBER': ['sum', 'mean'], 'NOVEMBER': ['sum', 'mean'], 'DECEMBER': ['sum', 'mean']}).round(0)
```
